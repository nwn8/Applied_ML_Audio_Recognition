This notebook will create a dataframe from the songs_database.db that will be used in the Machine Learning model.

## 1. Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3

# KNN model 
from sklearn.neighbors import NearestNeighbors
```

## 2. Load Data

```python
# The features will only be loaded from the Audio_features table and songs table.  Fingerprints table will not be used in the ML process.

# Connect to your SQLite database
conn = sqlite3.connect("songs_database.db")

# Load each table into a DataFrame
songs_df = pd.read_sql_query("SELECT song_id, word_count, unique_word_count FROM songs", conn)

audio_features_df = pd.read_sql_query("SELECT * FROM audio_features", conn)

# check dataframes loaded correctly

print(songs_df.head())
print(audio_features_df.head())


# Close connection when done
conn.close()
```

## 3. Data Transformation and Feature Engineering

Create the dataframe for the ML model by combining the audio features numerical data with the duration, word count, and unique word count from the songs table.

```python
audio_features_df = audio_features_df.drop(columns=["feature_id"])
combined_df = songs_df.merge(audio_features_df, on="song_id", how="inner")

print(combined_df.info())
```

### Final ML Features Dataset

combined_df still has a song_id identifier after combining.  Dropping it will create a numerical dataset to build the ML model

```python
ml_features_df = combined_df.drop(columns=["song_id"])

print(ml_features_df.info())
```
